### Now its for us to redirect the crawler to the profile we need to scrap

```python
visitingProfileID = '/in/shantanu-arya/'
fullLink = 'https://www.linkedin.com' + visitingProfileID
browser.get(fullLink)
```

#### A scroll of the entire page is needed just so that we are able to scrap the entire page and nothing is left out

```python
SCROLL_PAUSE_TIME = 5

# Get scroll height

last_height = browser.execute_script("return document.body.scrollHeight")

for i in range(3):
    # Scroll down to bottom
    browser.execute_script("window.scrollTo(0, document.body.ScrollHeight);")
    
    # Wait for the page to load
    time.sleep(SCROLL_PAUSE_TIME)
    
    # Calculate new scroll height and compare with last scroll height
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    
```

```python
src = browser.page_source
soup = BeautifulSoup(src,'lxml')
soup
```

### HEADER & ABOUT SECTION

```python
name_div = soup.find("div",{"class":"flex-1 mr5"})
name_div
```

```python
name_loc = name_div.find_all('ul')
print(name_loc)
name = name_loc[0].find('li').get_text().strip()
print("Name : ",name)
loc = name_loc[1].find('li').get_text().strip()
print("Location : ",loc)
```

```python
profile_title = name_div.find('h2').get_text().strip()
profile_title
```

```python
connection = name_loc[1].find_all('li')
connections = connection[1].get_text().strip()
connections
```

```python
info = []
info.append(fullLink)
info.append(name)
info.append(profile_title)
info.append(loc)
info.append(connections)
info
```
